```python
!pip install scikit-fuzzy
import skfuzzy as fuzz
import matplotlib.pyplot as plt
import numpy as np
import skfuzzy.membership as mf
from skfuzzy import control as ctrl

#Define the intervals for the input and output variables

x_assists= np.arange(0,301,1)
x_Fouls= np.arange(0,41,1)
x_Goals= np.arange(0,201,1)
x_PlayTime= np.arange(0,131,1)
x_YellowCard=np.arange(0,51,1)
x_RedCard=np.arange(0,21,1)
Salary= np.arange(0, 201, 1)

#Our Testing Data

input_yellowc=15
input_red_card= 4
input_assists=80
input_fouls= 15
input_goals= 150
input_playtime= 82


#Define the membership function for fuzzy logic

x_assists_low=mf.trimf(x_assists,[0, 50, 100])
x_assists_med=mf.trimf(x_assists,[50, 150, 250])
x_assists_high=mf.trimf(x_assists,[150, 250, 300])

x_Fouls_low=mf.trimf(x_Fouls,[0, 10, 20])
x_Fouls_med=mf.trimf(x_Fouls,[10, 20, 30])
x_Fouls_high=mf.trimf(x_Fouls,[20, 30, 40])

x_Goals_low=mf.trimf(x_Goals,[0, 50, 100])
x_Goals_med=mf.trimf(x_Goals,[50, 100, 150])
x_Goals_high=mf.trimf(x_Goals,[100, 150, 200])

x_PlayTime_low=mf.trimf(x_PlayTime,[0, 30, 60])
x_PlayTime_med=mf.trimf(x_PlayTime,[30, 60, 90])
x_PlayTime_high=mf.trimf(x_PlayTime,[60, 90, 130])

x_YellowCard_low=mf.trimf(x_YellowCard,[0, 15, 30])
x_YellowCard_med=mf.trimf(x_YellowCard,[15, 30, 45])
x_YellowCard_high=mf.trimf(x_YellowCard,[30, 45, 50])

x_RedCard_low=mf.trimf(x_RedCard,[0, 5, 10])
x_RedCard_med=mf.trimf(x_RedCard,[5, 10, 15])
x_RedCard_high=mf.trimf(x_RedCard,[10, 15, 20])

Salary_low= mf.trimf(np.arange(0, 201, 1), [0, 50, 100])
Salary_med= mf.trimf(np.arange(0, 201, 1), [50,100, 150])
Salary_high= mf.trimf(np.arange(0, 201, 1), [100, 150, 200])

#Let us show the figure for the membership functions

fig, (ax0, ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=6, figsize=(10, 40))
ax0.plot(x_Goals,x_Goals_low,'r',label= "Dusuk")
ax0.plot(x_Goals,x_Goals_med,'g',label= "Orta")
ax0.plot(x_Goals,x_Goals_high,'y',label= "Yuksek")
ax0.set_title("Futbolcunun Gol sayisi")
ax0.legend()

ax1.plot(x_Fouls,x_Fouls_low,'r',label= "Dusuk")
ax1.plot(x_Fouls,x_Fouls_med,'g',label= "Orta")
ax1.plot(x_Fouls,x_Fouls_high,'y',label= "Yuksek")
ax1.set_title("Futbolcunun yedigi faul sayisi")
ax1.legend()


ax2.plot(x_YellowCard,x_YellowCard_low,'r',label= "Dusuk")
ax2.plot(x_YellowCard,x_YellowCard_med,'g',label= "orta")
ax2.plot(x_YellowCard,x_YellowCard_high,'y',label= "yuksek")
ax2.set_title("Futbolcunun sahip oldugu sari kart sayisi")
ax2.legend()

ax3.plot(x_RedCard, x_RedCard_low, 'r', label="Dusuk")
ax3.plot(x_RedCard, x_RedCard_med, 'g', label="orta")
ax3.plot(x_RedCard, x_RedCard_high, 'y', label="yuksek")
ax3.set_title("Futbolcunun sahip oldugu kirmizi kart sayisi")
ax3.legend()

ax4.plot(x_assists, x_assists_low, 'r', label="Dusuk")
ax4.plot(x_assists, x_assists_med, 'g', label="orta")
ax4.plot(x_assists, x_assists_high, 'y', label="yuksek")
ax4.set_title("Futbolcunun yaptigi asist sayisi")
ax4.legend()

ax5.plot(x_PlayTime, x_PlayTime_low, 'r', label="Dusuk")
ax5.plot(x_PlayTime, x_PlayTime_med, 'g', label="orta")
ax5.plot(x_PlayTime, x_PlayTime_high, 'y', label="yuksek")
ax5.set_title("Futbolcunun bir macta oynadigi sure")
ax5.legend()


#Ä°nterpretation for each input values

YellowCard_fit_low=fuzz.interp_membership(x_YellowCard,x_YellowCard_low,input_yellowc)
YellowCard_fit_med=fuzz.interp_membership(x_YellowCard,x_YellowCard_med,input_yellowc)
YellowCard_fit_high=fuzz.interp_membership(x_YellowCard,x_YellowCard_high,input_yellowc)

RedCard_fit_low = fuzz.interp_membership(x_RedCard, x_RedCard_low, input_red_card)
RedCard_fit_med = fuzz.interp_membership(x_RedCard, x_RedCard_med, input_red_card)
RedCard_fit_high = fuzz.interp_membership(x_RedCard, x_RedCard_high, input_red_card)


Assists_fit_low = fuzz.interp_membership(x_assists, x_assists_low, input_assists)
Assists_fit_med = fuzz.interp_membership(x_assists, x_assists_med, input_assists)
Assists_fit_high = fuzz.interp_membership(x_assists, x_assists_high, input_assists)


Fouls_fit_low = fuzz.interp_membership(x_Fouls, x_Fouls_low, input_fouls)
Fouls_fit_med = fuzz.interp_membership(x_Fouls, x_Fouls_med, input_fouls)
Fouls_fit_high = fuzz.interp_membership(x_Fouls, x_Fouls_high, input_fouls)

Goals_fit_low = fuzz.interp_membership(x_Goals, x_Goals_low, input_goals)
Goals_fit_med = fuzz.interp_membership(x_Goals, x_Goals_med, input_goals)
Goals_fit_high = fuzz.interp_membership(x_Goals, x_Goals_high, input_goals)

PlayTime_fit_low = fuzz.interp_membership(x_PlayTime, x_PlayTime_low, input_playtime)
PlayTime_fit_med = fuzz.interp_membership(x_PlayTime, x_PlayTime_med, input_playtime)
PlayTime_fit_high = fuzz.interp_membership(x_PlayTime, x_PlayTime_high, input_playtime)


#Set the rules in a different ways

rule1 = np.fmin(np.fmax(Goals_fit_low, Assists_fit_high), Salary_high)
rule2 = np.fmax(np.fmax(Goals_fit_high, PlayTime_fit_med), Salary_med)
rule3 = np.fmin(np.fmax(YellowCard_fit_low, RedCard_fit_high), Salary_low)
rule4 = np.fmax(np.fmin(Fouls_fit_high, np.fmax(Assists_fit_low, Goals_fit_high)), Salary_med)
rule5 = np.fmin(np.fmax(PlayTime_fit_high, RedCard_fit_low), Salary_high)
rule6 = np.fmax(np.fmin(YellowCard_fit_low, np.fmin(RedCard_fit_high, Assists_fit_high)), Salary_low)
rule7 = np.fmin(np.fmax(Assists_fit_high, Fouls_fit_high), Salary_med)
rule8 = np.fmax(np.fmin(Goals_fit_low, np.fmax(PlayTime_fit_med, RedCard_fit_low)), Salary_high)
rule9 = np.fmin(np.fmax(YellowCard_fit_low, np.fmax(Assists_fit_low, Goals_fit_high)), Salary_high)
rule10 = np.fmax(np.fmax(Fouls_fit_high, np.fmin(Goals_fit_low, RedCard_fit_high)), Salary_low)

#Combing the rules for one output (Deffuziation)

Salary_cikti_high = np.fmin(rule1, np.fmax(rule5, np.fmax(rule8, rule9)))
Salary_cikti_med = np.fmin(rule2, np.fmax(rule4, rule7))
Salary_cikti_low = np.fmin(rule3, np.fmax(rule6, rule10))


Salary_cikti=np.fmax(Salary_low,np.fmax(Salary_med,Salary_high))
result=fuzz.interp_membership(Salary,Salary_cikti,Salary_cikti)
print(result)


# Visualization
Salary0=np.zeros_like(Salary)
fig,ax0=plt.subplots(figsize=(7,4))
ax0.fill_between(Salary,Salary0,Salary_low,color="r",alpha=0.7)
ax0.plot(Salary,Salary_low,'g')
ax0.fill_between(Salary,Salary0,Salary_med,color='g',alpha=0.7)
ax0.plot(Salary,Salary_med,'r')
ax0.fill_between(Salary,Salary0,Salary_high,color='y',alpha=0.7)
ax0.plot(Salary,Salary_high,'b')
ax0.set_title('Salary Membership and Defuzzification')
ax0.legend()
plt.show()


Salary_cikti=np.fmax(Salary_low,np.fmax(Salary_med,Salary_high))
deffuzified=fuzz.defuzz(Salary,Salary_cikti,'centroid')
result=fuzz.interp_membership(Salary,Salary_cikti,deffuzified)

#Result
```
